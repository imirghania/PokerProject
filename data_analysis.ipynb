### Players Performace Analysis

```python
def analyze_player_performance(df:pd.DataFrame, player_name:str):
    """This function takes a Pandas dataframe of pocker hands, each row in the dataframe
        represents one hand. The function extracts the hands related to the provided `player_name`
        Parameters:
            df (pd.DataFrame): a dataframe of Holdem poker hands.
            player_name (str): a player name
        returns:
            dict: a dict of information related to the player_name, either directly extracted from the 
                  dataframe hands or calculated based on some columns.
    """
    result = {}
    player_hands_df = df_utils.get_player_hands(df, player_name)
    result_df = player_hands_df.assign(
        is_hand_won = lambda df_: df_.apply(df_utils.hand_winning_test, 
                                            player_name=player_name, 
                                            axis=1),
        aggression_factor = lambda df_: df_utils.calculate_aggression_factor(df_, player_name),
        player_total_bets = lambda df_: df_.apply(df_utils.get_player_hand_bet, player_name=player_name, axis=1),
        player_hand_profit = lambda df_: df_.apply(df_utils.calculate_hand_profit, player_name=player_name, axis=1),
        player_hand_bankroll = lambda df_: df_.apply(df_utils.get_player_hand_bankroll, player_name=player_name, axis=1)
    )
    result["player_name"] = player_name
    result["hands_count"] = result_df.shape[0]
    result["hands_won_count"] = result_df[result_df.is_hand_won].shape[0]
    hands_won = result_df[result_df.is_hand_won]
    hands_lost = result_df[~result_df.is_hand_won]
    result["win_aggression_factor"] = round(hands_won["aggression_factor"].mean().item(), 3) if hands_won.shape[0] else 0
    result["loss_aggression_factor"] = round(hands_lost["aggression_factor"].mean().item(), 3) if hands_lost.shape[0] else 0
    result["actions"] = result_df["player_total_bets"].sum().item()
    result["profit"] = result_df["player_hand_profit"].sum().item()
    result["return"] = round( (result["actions"]+result["profit"])/result["actions"], 3)
    result["SB_per_hand"] = round( result["profit"]/result["hands_count"], 3)
    result["bankroll_mean"] = round( result_df["player_hand_bankroll"].mean().item(), 3)
    result["bankroll_std"] = round( result_df["player_hand_bankroll"].std().item(), 3)


    return result
```

```python
analyze_player_performance(df, 'boris')
```

```python
def analyze_players_performaces(hands_df:pd.DataFrame, players:list[str]):
    """This function takes a Pandas dataframe of pocker hands, each row in the dataframe
        represents one hand. The function analyzes the performance of each player whose name is provided
        in `players` list of players names.
        Parameters:
            df (pd.DataFrame): a dataframe of Holdem poker hands.
            players (list[str]): a list of players names
        returns:
            list[dict]: a list of dicts  each contains information corresponding to a player name 
            from `players` either directly extracted from the dataframe hands or calculated based 
            on some columns.
    """
    performace_list = []
    for player_name in tqdm(players, desc="Analyzing players performance", total=len(players), unit='player'):
        player_performace = analyze_player_performance(hands_df, player_name)
        performace_list.append(player_performace)
    return performace_list
```

```python
players_reports = analyze_players_performaces(df, players_names)
```

### Constructing Peformance DataFrames

```python
def merge_player_performance_reports(list_of_dicts:list[dict]):
    combined_dict = {}
    for d in list_of_dicts:
        for key, value in d.items():
            combined_dict.setdefault(key, []).append(value)
    return combined_dict
```

```python
merged_players_reports = merge_player_performance_reports(players_reports)
```

```python
players_reports_df = pd.DataFrame(merged_players_reports)
```
